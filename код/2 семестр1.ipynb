```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - 2*x**2 - 11*x + 12

def bisection_method(a, b, tolerance, max_iterations):
    iterations = 0
    while abs(b - a) > tolerance and iterations < max_iterations:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return (a + b) / 2

# Интервалы поиска корней
intervals = [(-4, -2), (-2, 0), (2, 4)]

# Начальные приближения для графика
initial_guesses = [(-4.5, 0), (-2.5, 0), (2.5, 0)]

# Решение уравнения и построение графика
plt.figure()
x = np.linspace(-5, 5, 1000)
plt.plot(x, f(x), label='x^3-2x^2-11x+12')
for interval, initial_guess in zip(intervals, initial_guesses):
    root = bisection_method(interval[0], interval[1], 0.0001, 100)
    plt.plot(root, 0, marker='x', color='red')
    plt.annotate(f'({root:.2f}, 0)', (root, 0), textcoords="offset points", xytext=initial_guess, ha='center')
    print(f'Root: {root}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()




#Результаты решения уравнения методом половинного деления:

#Root: -3.000091552734375
#Root: 1.9999542236328125

#Результаты совпадают со значениями, полученными при использовании функции с библиотеки Python.
```

```python
import matplotlib.pyplot as plt
import numpy as np # Импортируе необходимые библиотеки: matplotlib.pyplot и numpy.

# Объявление функции, содержащей правую часть уравнения
def my_fun(I):
  E = 5
  r = 0.5
  return I - E/(r+np.exp(-0.001*I**2))
I = np.arange(1,5,0.01) # Создание массивов координат точек графика
Y = my_fun(I) # создаём  массивы значений `I`, используя `np.arange`, и вычисляет соответствующие значения `Y`, используя функцию `my_fun`
# Построение графика
plt.figure(1)
plt.plot(I,Y)
plt.grid()

from scipy.optimize import bisect #импортируем функцию `bisect` из библиотеки scipy.

I = bisect(my_fun,3,3.5) #  функция `bisect` используется для поиска корня функции `my_fun` в интервале [3, 3.5].

print('Расчёт тока схемы I = {:.2f}'.format(I)) #выводит вычисленное значение `I`, используя форматированную строку.




```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**3 - 2*x**2 - 11*x + 12

def f1(a, b, e):
    if f(a)*f(b) >= 0:
        print("Нет корня на выбранном интервале")
        return None

    while (b - a) / 2 > e:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2


# Находим корни уравнения
root1 = f1(-4, -2, 0.0001)
root2 = f1(0, 2, 0.0001)
root3 = f1(3, 5, 0.0001)

print("Корни уравнения:")
print("x1 =", root1)
print("x2 =", root2)
print("x3 =", root3)

# Строим график
x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^3 - 2x^2 - 11x + 12')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.scatter([root1, root2, root3], [0, 0, 0], color='red', label='Корни уравнения')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции и корней уравнения')
plt.show()
```
